### Load Amazon Data into Spark DataFrame

```python
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Books_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get("amazon_reviews_us_Books_v1_00.tsv.gz"), sep="\t", header=True, inferSchema=True)
df.show()
```

### Create DataFrames to match tables

```python
from pyspark.sql.functions import to_date
```

```python
# Read in the Review dataset as a DataFrame
```

```python
# Create the customers_table DataFrame
# customers_df = df.groupby("").agg({""}).withColumnRenamed("", "customer_count")
customers_df = df.groupby("customer_id").agg({"customer_id":"count"}).withColumnRenamed("count(customer_id)", "customer_count")
```

```python
pt(customers_df)
```

```python
customers_df.dtypes
```

```python
customers_df.count()
```

```python
customers_df.show()
```

```python
# Create the products_table DataFrame and drop duplicates. 
# products_df = df.select([]).drop_duplicates()
products_df = df.select(["product_id","product_title"]).drop_duplicates(["product_id"])
```

```python
pt(products_df)
```

```python
products_df.dtypes
```

```python
products_df.count()
```

```python
products_df.show()
```

```python
# Create the review_id_table DataFrame. 
# Convert the 'review_date' column to a date datatype with to_date("review_date", 'yyyy-MM-dd').alias("review_date")
# review_id_df = df.select([, to_date("review_date", 'yyyy-MM-dd').alias("review_date")])
review_id_df = df.select(["review_id", "customer_id", "product_id", "product_parent", to_date("review_date", 'yyyy-MM-dd').alias("review_date")])
```

```python
pt(review_id_df)
```

```python
review_id_df.dtypes
```

```python
review_id_df.count()
```

```python
review_id_df.show()
```

```python
from pyspark.sql import types
```

```python
# Create the vine_table. DataFrame
# vine_df = df.select([])
vine_df = df.select(["review_id", df["star_rating"].cast("int").alias("star_rating"), "helpful_votes", "total_votes", "vine", "verified_purchase"])
```

```python
pt(vine_df)
```

```python
vine_df.dtypes
```

```python
vine_df.count()
```

```python
vine_df.show()
```

### Connect to the AWS RDS instance and write each DataFrame to its table. 

```python
from getpass import getpass
```

```python
# Configure settings for RDS
mode = "append"
jdbc_url="jdbc:postgresql://module-16-challenge.crsqermsvhry.us-west-1.rds.amazonaws.com:5432/Amazon_Vine_Analysis"
password = getpass('Enter database password')
config = {"user":"postgres", 
          "password": password, 
          "driver":"org.postgresql.Driver"}
```

```python
# Write review_id_df to table in RDS
review_id_df.write.jdbc(url=jdbc_url, table='review_id_table', mode=mode, properties=config)
```

After upload, `review_id_table` in database shows 10319090 rows, which matches the count from above.

```python
# Write products_df to table in RDS
# about 3 min
products_df.write.jdbc(url=jdbc_url, table='products_table', mode=mode, properties=config)
```

After upload, `products_table` in database shows 2274543 rows, which matches the count from above.

```python
# Write customers_df to table in RDS
# 5 min 14 s
customers_df.write.jdbc(url=jdbc_url, table='customers_table', mode=mode, properties=config)
```

After upload, `customers_table` in database shows 4632497 rows, which matches the count from above.

```python
# Write vine_df to table in RDS
# 11 minutes
vine_df.write.jdbc(url=jdbc_url, table='vine_table', mode=mode, properties=config)
```

After upload, `vine_table` in database shows 10319090 rows, which matches the count from above.
